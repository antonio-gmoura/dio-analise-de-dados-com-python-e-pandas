```python
#importando a biblioteca
import pandas as pd
```

```python
#leitura dos arquivos Excel.
df1 = pd.read_excel("Datasets/Aracaju.xlsx")
df2 = pd.read_excel("Datasets/Fortaleza.xlsx")
df3 = pd.read_excel("Datasets/Natal.xlsx")
df4 = pd.read_excel("Datasets/Recife.xlsx")
df5 = pd.read_excel("Datasets/Salvador.xlsx")
```

```python
#juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#exibindo as 5 Ãºltimas linhas
df.tail()
```

```python
#alterando o tipo de dado da coluna LojaID, de nÃºmero para string, para nÃ£o haver cÃ¡lculos automÃ¡ticos sobre ela.
df["LojaID"] = df["LojaID"].astype("object")
```

```python
#verificando novamente o tipo de dado de cada coluna
df.dtypes
```

```python
#verificando novamente o dataframe
df.head()
```

## Tratando valores faltantes

```python
#consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#Substituindo os valores nulos pela mÃ©dia
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
#obtendo a mÃ©dia de vendas
df["Vendas"].mean()
```

```python
#checando a substituiÃ§Ã£o dos valores nulos.
df.isnull().sum()
```

```python
#Obtendo uma amostra de 15 registros.
df.sample(15)
```

```python
#substituindo os valores nulos por zero
df["Vendas"].fillna(0, inplace=True)
```

```python
#apagando as linhas com valores nulos
df.dropna(inplace=True)
```

```python
#apagando as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

## Criando colunas novas

```python
#Criando a coluna de receita (venda * qtde)
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#obtendo a quantidade pela engenharia reversa
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
df.head()
```

```python
#Retornando a maior receita
df["Receita"].max()
```

```python
#Retornando a menor receita
df["Receita"].min()
```

```python
#nlargest (3 melhores receitas)
df.nlargest(3, "Receita")
```

```python
#nsamllest (3 piores receitas)
df.nsmallest(3, "Receita")
```

```python
#agrupamento por cidade, somando receita.
df.groupby("Cidade")["Receita"].sum()
```

```python
#ordenando o conjunto de dados
df.sort_values("Receita", ascending=False).head(10)
```

# *Aula 5 - Trabalhando com datas*

```python
#trasnformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
```

```python
#verificando o tipo de dado de cada coluna
df.dtypes
```

```python
#transformando coluna de data em tipo data
df["Data"] = pd.to_datetime(df["Data"])
```

```python
#verificando o tipo de dado de cada coluna
df.dtypes
```

```python
#agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#criando uma nova coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
#obtendo 5 amostras
df.sample(5)
```

```python
#extraindo o mÃªs e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
#obtendo 5 amostras
df.sample(5)
```

```python
#retornando a data mais antiga
df["Data"].min()
```

```python
#calculando a diferenÃ§a de dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
```

```python
#obtendo 5 amostras
df.sample(5)
```

```python
#criando a coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
```

```python
#obtendo 5 amostras
df.sample(5)
```

```python
#criando um dataframe com as vendas de 2019 do mÃªs de marÃ§o
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```

```python
#obtendo 20 amostras do novo dataframe
vendas_marco_19.sample(20)
```
